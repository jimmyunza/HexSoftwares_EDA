# EXPLORATIVE DATA ANALYSIS for Iris Dataset

```python
# importing packages

import pandas as pd
from sklearn import datasets
import seaborn as sns
import matplotlib.pyplot as plt

# loading iris dataset as numPy array

iris = datasets.load_iris()
```

- From the code above it desribes the characheristices of the datasets
- The dataset has 150 rows (150 samples), belonging to 3 classes/ categories which are Setosa,     Virginica and Versicolour.
- The dataset comprorise of four(4) (features) which are sepal length, sepal width, petal length and petal width in cm respectively.
- The data is balanced with 50 samples for each class
- Data has no missing values
- Petal dimensions correlates highly to the classes more than sepal dimensions 


```python
# loading dataset to dataFrame
"""
#loading data as DataFrame using pandas library to have a clear and easy read and easy data exploration

df = pd.DataFrame(iris.data, columns= irei.feature_names)
"""
df = pd.DataFrame(iris.data, columns= iris.feature_names)
df["target"] = pd.Series(iris.target)
df['target_names'] = df['target'].apply(lambda y: iris.target_names[y])

# print head of data frame
df.sample(5).head()
```

```python
# Plots for each feature

# select features
features = ["sepal length (cm)","sepal width (cm)", "petal length (cm)", "petal width (cm)"]
plt.figure(figsize= (10 ,6))
sns.boxplot(data=df[features], linewidth=2,)
plt.title("Features Boxplots")
plt.show()
```

- More than 75% of the dataset sepal length is above 5cm, with a median of around 5.8cm. The maximum and minimum sepal length is around 7.9cm and 4.3cm respectively.
- For sepal width, the length ranges from about 2.2cm to 4cm with 50% sepal width is >=3cm. There are few outliers above and below the maximum value and minimum value
- For petal length 50%, is between 1cm and 4.3cm. The maximum length is <=7cm
- From the box plots, there is a positive correlation between sepal length and petal length meaning as petal length increases, sepal length also increases

```python
# Box Plot by each feature
plt.figure(figsize=(10,6))
sns.boxplot(data= df, x='target_names', y = 'sepal length (cm)', linewidth=2,)
plt.title("Box Plot for Sepal Length by Specie")
plt.show()
```

- From the plots Setosa, Versicolor and Virginica has a minimum sepal length of 3.5cm, 4.9cm and 5.7cm and, a maximum length of 5.8cm, 7cm and 7.9cm respectively which shows it is difficult to seperate the classes using sepal length

- From the dataset it is clearly difficult to seperate the classes by sepal length as each class overlaps the other classes as shown by the plots

```python
# box plot  by sepal width for each specie
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x= 'target_names', y="sepal width (cm)", linewidth=2,)
plt.title("Box Plot By Sepal Width")
plt.show()
```

From the boxplots, Setosa have longer sepal width than the other species, with 50% of sepal width above 3.4cm.
However, it is difficulty to seperate each class as sepal length overlaps in every class

```python
#  box plot for petal by lenght
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, linewidth=2, y='petal length (cm)', x= 'target_names')
plt.title('Petal Length by Specie')
plt.show()
```

Setosa petal length are very short the the other 2 classes. However, versicolor petal length above the median length, also falls in the first quarter of Virginica 

From the box plots, setosa are easily separated from the other classes versicolor and virginica that have petal lenght overlap

```python
# petal width box plots by species
plt.figure(figsize=(10,6))
sns.boxplot(linewidth=2, data=df, x='target_names', y='petal width (cm)')
plt.title('Box Plots for Petal Width by Spicies')
plt.show()
```

Setosa petal width are very short the the other 2 classes. However, versicolor petal width above the median length, also falls in the first quarter of Virginica.

Also setosa are seperated from virginica and versicolor and the petal width overlaps in those classes

```python
################# Exploration using scatter plots #############
# crrate figure with muiltipl plots

fig, axs = plt.subplots(3, 2, figsize=(12,12))

# sepal length vs sepal width
axs[0,0].scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=df['target'])
axs[0,0].set_title('fig.1: Sepal Length Vs Sepal Width')
axs[0,0].set_xlabel("Sepal Length (cm)")
axs[0,0].set_ylabel('Sepal Width (cm)')

# sepal length vs petal length
axs[0,1].scatter(df['sepal length (cm)'], df['petal length (cm)'], c=df['target'])
axs[0,1].set_title('fig.2: Sepal Length Vs Petal Length')
axs[0,1].set_xlabel('Sepal Length (cm)')
axs[0,1].set_ylabel('Petal Length (cm)')

# sepal width vs petal Width
axs[1,0].scatter(df['sepal width (cm)'], df['petal width (cm)'], c=df['target'])
axs[1,0].set_title('fig.3: Sepal Width Vs Petal Width')
axs[1,0].set_xlabel('Sepal Width (cm)')
axs[1,0].set_ylabel('Petal Width (cm)')

# petal length vs sepal width
axs[1,1].scatter(df['sepal width (cm)'], df['petal length (cm)'], c=df['target'])
axs[1,1].set_title("fig.4: Sepal Width Vs Petal Length")
axs[1,1].set_xlabel("Sepal Width (cm)")
axs[1,1].set_ylabel("Petal length (cm)")

# petal length vs Petal width
axs[2,0].scatter(df['petal length (cm)'], df['petal width (cm)'], c=df['target'])
axs[2,0].set_title("fig.5: Petal Length Vs Petal Width")
axs[2,0].set_xlabel("Petal Length (cm)")
axs[2,0].set_ylabel("Petal Width (cm)")

# sepal length vs Petal width
axs[2,1].scatter(df['sepal length (cm)'], df['petal width (cm)'], c=df['target'])
axs[2,1].set_title("fig.6: Sepal Length Vs Petal Width")
axs[2,1].set_xlabel("Petal Length (cm)")
axs[2,1].set_ylabel("Petal Width (cm)")

#add a legend
for i in axs.flat:
    i.legend(df.target_names, loc='upper right', bbox_to_anchor=(1.2,1), fontsize='small')

# Display Plots
plt.tight_layout()
plt.show()
```

- Versicolor (Green dots) and Virginica (Yellow dots) are more overlapping, making it less distinguishable using sepal length and sepal Width. These species have larger Sepal length and smaller sepal width compareed to setosa but the ranges overlap.
- This plot suggest that Setosa is linearly seperable from the other two classes based on these two features
